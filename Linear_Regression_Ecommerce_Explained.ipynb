```python
# Splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
# General info about the dataset (columns, types, nulls, etc.)
df.info()
```

```python
X=df[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y=df['Yearly Amount Spent']
```

```python
# Splitting the dataset into training and testing sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# Importing Linear Regression model from sklearn
from sklearn.linear_model import LinearRegression
```

```python
# Importing Linear Regression model from sklearn
lm = LinearRegression()
```

```python
# Fitting the linear model on training data
lm.fit(X_train,y_train)
```

```python
lm.coef_
```

```python
# Making predictions using the trained model
pred=lm.predict(X_test)
```

```python
# Plotting scatter plot to visualize relationship between features and target
sns.scatterplot(x=y_test,y=pred)
# Label for y-axis
plt.ylabel("Predictions")
# Adding a title to the plot
plt.title("Actual vs Predictions")
```

```python
# Calculating Mean Squared Error for evaluation
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
print('Mean Absolute Error:',mean_absolute_error(y_test, pred))
# Calculating Mean Squared Error for evaluation
print('Mean Squared Error:',mean_squared_error(y_test, pred))
# Calculating Mean Squared Error for evaluation
print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, pred)))
```
